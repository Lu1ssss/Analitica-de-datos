**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**



```python
#filtro por comparacion
filtro6=df[df["population"] > 8000]
filtro6
```

**FILTROS BASICOS CON CONECTORES LOGICOS**

```python
#filtro Y
filtro7=df[(df["population"] > 8000) & (df["continent"] == "Africa")]
filtro7
```

```python
#filtro O
filtro8=df[(df["population"] > 8000) | (df["continent"] == "Africa")]
filtro8
```

```python
#filtro not
filtro9=df[~(df["population"] >= 8000) & ~(df["continent"] == "Africa")]
filtro9
```
